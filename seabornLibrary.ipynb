# Line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [8,7,6,5,4,3,2,1]

#matplotlib
plt.plot(x,y)
plt.title("matplotlib")
plt.show()
```

```python
a = [1,2,3,4,5,6,7,8]
b = [8,7,6,5,4,3,2,1]

# Seaborn
sns.lineplot(x = a , y= b)
plt.title("Seaborn")
plt.show()
```

```python
a = [1,2,3,4,5,6,7,8]
b = [8,7,6,5,4,3,2,1]

import pandas as pd

df = pd.DataFrame({'var' : a, 'var1' : b})
sns.lineplot(x = 'var', y= 'var1' , data =df)
plt.show()
```

# Bar plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = sns.load_dataset("penguins")
df1
```

```python
sns.barplot(x=df1.island , y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island" ,y = "bill_length_mm" , data = df1)
plt.show()
```

categorical data based on male and female

```python
sns.barplot(x = "island" ,y = "bill_length_mm" , data = df1, hue="sex")
plt.show()
```

ordering ourself using order parameter

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island" ,y = "bill_length_mm" , data = df1, hue="sex" , order = order1)
plt.show()
```

hue_order parameter for changing order of bar

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island" ,y = "bill_length_mm" , data = df1, hue="sex" , order = order1, hue_order = ["Female","Male"])
plt.show()
```

```python
sns.barplot(x = "island" ,y = "bill_length_mm" , data = df1, color = "red")
plt.show()
```

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island" ,y = "bill_length_mm" , data = df1, palette = "gist_earth", hue="sex" , order = order1)
plt.show()
```

Saturation parameter to fade color

```python
sns.barplot(x = "island" ,y = "bill_length_mm" , data = df1, palette = "gist_heat_r", hue="sex" , saturation = 0.5)
plt.show()
```

histogram

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

bins parameter for proper intervals

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240])
plt.show()
```

color

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240] , color = "red")
plt.show()
```

kde parameter to show density of kernel

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240] , kde = True)  # kde is the line drawn in the histrogram
```

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240] , kde= True , color = "green")
plt.show()
```
