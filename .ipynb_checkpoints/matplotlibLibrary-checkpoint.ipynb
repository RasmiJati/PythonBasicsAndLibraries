```python
import matplotlib.pyplot as plt
```

# Linear graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'g'   # change color of line
# plt.plot(x,y)
plt.plot(x,y,c)
plt.show()  # only to display graph
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
plt.plot(y,marker='>')
plt.plot(x,marker='o')
plt.show()
```

# Bar graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

# change color of each bar

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y,color = c)
plt.show()
```

# using array

```python
import numpy as np
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
c = ['r','g','b','y']
plt.bar(x,y,color=c)
```

# horizontal representation

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
c = ['r','g','b','y']
plt.barh(x,y,color=c)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
# plt.bar(x,y,color="red")
plt.bar(x,y,color="#4CAF50")
plt.show()
```

# provide width for vertical and height for horizontal

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.bar(x,y,color="#4CAF50" , width = 0.3)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.barh(x,y,color="#4CAF50" , height = 0.3)
plt.show()
```

# Scatter plot

```python
x =[1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.scatter(x,y)
plt.show()
```

# to give title to the plot

```python
x =[1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.scatter(x,y)
plt.show()
```

# give label to axis

```python
x =[1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.xlabel("Month")
plt.ylabel("Numbers")
c = ['red','green','yellow','purple','black','orange']
plt.scatter(x,y, color = c)
#plt.scatter(x,y)
# plt.scatter(x,y, color = "red")
plt.show()
```
