# Bar graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

# change color of each bar

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y,color = c)
plt.show()
```

# using array

```python
import numpy as np
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
c = ['r','g','b','y']
plt.bar(x,y,color=c)
```

# horizontal representation

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
c = ['r','g','b','y']
plt.barh(x,y,color=c)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
# plt.bar(x,y,color="red")
plt.bar(x,y,color="#4CAF50")
plt.show()
```

# provide width for vertical and height for horizontal

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.bar(x,y,color="#4CAF50" , width = 0.3)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.barh(x,y,color="#4CAF50" , height = 0.3)
plt.show()
```

# Scatter plot

```python
x =[1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.scatter(x,y)
plt.show()
```

# to give title to the plot

```python
x =[1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.scatter(x,y)
plt.show()
```

# give label to axis

```python
x =[1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.xlabel("Month")
plt.ylabel("Numbers")
c = ['red','green','yellow','purple','black','orange']
plt.scatter(x,y, color = c)
#plt.scatter(x,y)
# plt.scatter(x,y, color = "red")
plt.show()
```

# change size of the plot

```python
x =[1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.xlabel("Month")
plt.ylabel("Numbers")
c = ['red','green','yellow','purple','black','orange']
plt.scatter(x,y, color = c, s = 250)
plt.show()
```

# changing color of image using matplotlib

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname = r'download.png'

#opening image using PIL
image = Image.open(fname).convert('L')

#mapping image to gray scale
plt.imshow(image, cmap = 'gray')
plt.show()
```

```python
bird = r'bird.jpg'
image = Image.open(bird).convert('L')
plt.imshow(image, cmap='gray')
plt.show()
```

```python
bird = r'bird.jpg'
image = Image.open(bird).convert('L')
plt.imshow(image, cmap='winter')
plt.show()
```

# to get exact original image convert from L to RGB or remove cmap

```python
bird = r'bird.jpg'
image = Image.open(bird).convert('RGB')
plt.imshow(image, cmap='winter')
plt.show()
```

```python
bird = r'person.jpg'
image = Image.open(bird).convert('L')
plt.imshow(image, cmap='gray')
plt.title("salman")
plt.show()
```

# to create grid

```python
bird = r'person.jpg'
image = Image.open(bird).convert('L')
plt.imshow(image, cmap='gray')
plt.title("salman")
plt.grid()
plt.show()
```

# Pie chart

```python
x = [10,20,30,40]
y = ['eng','nep','maths','sci']
plt.pie(x, labels =y)
plt.show()
```

# change to our own color

```python
x = [10,20,30,40]
y = ['eng','nep','maths','sci']
c = ['red','purple','pink','yellow']
plt.pie(x, labels =y, colors = c)
plt.show()
```

# create index label

```python
x = [10,20,30,40]
y = ['eng','nep','maths','sci']
c = ['red','purple','pink','yellow']
plt.pie(x, labels =y, colors = c)
plt.legend()
plt.show()
```
